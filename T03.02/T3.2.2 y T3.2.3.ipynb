```python
import pandas as pd
import numpy as np
df = pd.read_csv('DelayedFlights.csv')

print ('Resumeix estadísticament el dataset i les columnes d’interès. Fes una anàlisi estadístic del que consideris rellevant.')

# numero de vuelos por mes / mes con mas retrasos
print (df.groupby('Month')['ArrDelay'].count())
#-media de retrasos por mes
print (df.groupby('Month')['ArrDelay'].mean())
#-dia de la semana con mas retrasos
print (df.groupby('DayOfWeek')['ArrDelay'].count())
#3 origenes con mas retrasos a la salida
print (df.groupby('Origin')['DepDelay'].count().nlargest(3))
#3 destinos con mas retrasos a la llegada
print (df.groupby('Dest')['ArrDelay'].count().nlargest(3))


print ('Troba quantes dades faltants hi ha per columna.')
print (df.isnull().sum())

print ('Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).')
#velocidad media
df['AveSpe']=((df['Distance'])/(df['AirTime']))
#tarde o no
df['ArrLate'] = np.where(df['ArrDelay']>0,True, False)
print (df)

print('Fes una taula de les aerolínies amb més endarreriments acumulats.')
tabla = (df.groupby('UniqueCarrier')['ArrDelay'].count().nlargest(10))
print (tabla)

print('Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.')
maslargosdf = df.loc[:,['Origin', 'Dest', 'AirTime']]
print(maslargosdf.nlargest(5,'AirTime'))
masretasadodf = df.loc[:,['Origin', 'Dest', 'ArrDelay']]
print(masretasadodf.nlargest(5,'ArrDelay'))

print('Aporta allò que consideris rellevant.')
print('Se muestran los 10 destinos en los que más tardan de media los aviones en salir de pista y los 10 en los que menos')
print (df.groupby('Dest')['TaxiOut'].mean().nlargest(10))
print (df.groupby('Dest')['TaxiOut'].mean().nsmallest(10))

print('TASCA 3')
print('Exporta el dataset net i amb les noves columnes a Excel.')
df.to_excel('myDataFrame.xlsx')
#da error por el tamaño del fichero
```
